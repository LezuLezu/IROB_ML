```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w = 2*np.pi
t = np.linspace(0, 1.0, num=100)
```

```python
y1 = np.sin(w*t)
plt.plot(t,y1)
```

```python
y3 = 1/3*np.sin(3*w*t)
plt.plot(t,y3)
```

```python
y5 = 1/5*np.sin(5*w*t)
plt.plot(t,y5)
```

```python
y7 = 1/7*np.sin(7*w*t)
plt.plot(t,y7)
```

```python
# fourier series
fourier = y1+y3+y5+y7
plt.plot(t,fourier)
```
