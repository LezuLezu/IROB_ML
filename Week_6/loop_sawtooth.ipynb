```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
w = 2*np.pi
t = np.linspace(0, 1.0, num=100)
```

```python
# loop to create 100+ sawtooth waves in a plot each with a different frequency
x = 100
fourier = 0
for i in range(1, x+1):
    y = 1 / i * np.sin(i * w * t)
    fourier += y
plt.plot(t, fourier)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title("fourier with %s waves" % x)
plt.axline((0, 0), slope=0, color='k')
plt.show()
```
