```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
w = 2*np.pi
t = np.linspace(0, 1.0, num=100)
```

```python
y1 = np.sin(w*t)
plt.plot(t,y1)
```

```python
# make sawtooth wave of y1
plt.plot(t, signal.sawtooth(w * t))
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title("sawtooth y1")
plt.axline((0, 0), slope=0, color='k')
plt.show()
```

```python
y3 = 1/3*np.sin(3*w*t)
plt.plot(t,y3)
```

```python
# sawtooth wave of y3
plt.plot(t, signal.sawtooth(3 * w * t))
plt.xlabel('Time')
plt.ylabel('Amplitude') 
plt.title("sawtooth y3")
plt.axline((0, 0), slope=0, color='k')
plt.show()
```

```python
y5 = 1/5*np.sin(5*w*t)
plt.plot(t,y5)
```

```python
# Sawtooth wave of y5
plt.plot(t, signal.sawtooth(5 * w * t))
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title("sawtooth y5")
plt.axline((0, 0), slope=0, color='k')
plt.show()
```

```python
y7 = 1/7*np.sin(7*w*t)
plt.plot(t,y7)
```

```python
# Sawtooth wave of y7
plt.plot(t, signal.sawtooth(7 * w * t))
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title("sawtooth y7")
plt.axline((0, 0), slope=0, color='k')
plt.show()
```

```python
# fourier series
fourier = y1+y3+y5+y7
plt.plot(t,fourier)
```

```python
y2 = 1/2*np.sin(2*w*t)
y4= 1/4*np.sin(4*w*t)
y6 = 1/6*np.sin(6*w*t)
y8 = 1/8*np.sin(8*w*t)

fourier2 = y1+y2+y3+y4+y5+y6+y7+y8
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title("fourier 1 to 8")
plt.axline((0, 0), slope=0, color='k')
plt.plot(t,fourier2)
```
