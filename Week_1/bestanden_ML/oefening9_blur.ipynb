```python
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
```

```python
afbeelding = cv.imread('kodak dataset/kodim23.png')
```

```python
# convert to RGB
afbeelding = cv.cvtColor(afbeelding, cv.COLOR_BGR2RGB)
```

```python
plt.imshow(afbeelding)
plt.show()
```

```python
kernel = np.ones((3,3),np.float32)/9
kernel
```

```python
afbeelding = cv.filter2D(afbeelding,-1,kernel)
```

```python
plt.imshow(afbeelding)
plt.show()
```
